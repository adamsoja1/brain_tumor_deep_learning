```python
import os
import nibabel as nib
import numpy as np
```

```python
path = '../RSNA_ASNR_MICCAI_BraTS2021_TrainingData_16July2021'
files = os.listdir(path)
```

```python
def read_max_val(path,files):
    filex = []
    vals = []
   
    for file in files:
        img = nib.load(f'{path}/{file}/{file}_t1ce.nii.gz').get_fdata()
        image = img[:,:,60:115]
        
        image=np.array(image)
        
        temp = image.max()
        filex.append(file)
        vals.append(temp)
        
    return file,vals
        
        
        
    
    
        
    
    
```

```python
def image_load_nibabel(path,files,batch_size):
    L = len(files)
    while True:
        batch_start = 0
        batch_size_end = batch_size
        while batch_start < L:
            limit = min(batch_size_end,L)
            
            files_batched = files[batch_start:limit]
            
            #loading data
            x_train = []
            
            for file in files_batched:
                
                X_train = nib.load(f'{path}/{file}/{file}_t1ce.nii.gz').get_fdata()
                xyz = X_train[:,:,60:115]
                xyz = np.array(xyz)
                x_train.append(X_train)
                
            x_train = np.array(x_train)

           
            yield(x_train)
            
            batch_start +=batch_size
            batch_size_end +=batch_size
            
```

```python
batch_size = 50
x = image_load_nibabel(path,files,batch_size)
y = next(x)
y.max()
```
